## XGboost model

```python
import xgboost as xgb
```

```python
# Load data in XGboost format 'DMatrix'
D_train=xgb.DMatrix(X_train,label=y_train)
D_test=xgb.DMatrix(X_test,label=y_test)
```

```python
params = {
    'eta': 0.415,   # Learning rate
    'max_depth': 2,  
    'objective': 'reg:squarederror',  
    'gamma': 4.5,    #Minimum loss reduction required to make a further partition on a leaf node of the tree
    'lambda':0.08,  # L2 regularization
    'tree_method':'exact',  # Exact tree constructing algoritm for small datasets
    'eval_metric':'mae',
    'alpha':0.19,
    'seed':42,
    'min_child_weight':3    # minimum number of instances needed to be in each node
}

steps = 20  # The number of training iterations
```

```python
model = xgb.train(params, D_train, steps)
```

```python
preds = model.predict(D_test)
```

```python
preds
```

```python
d_test=pd.DataFrame(y_test)
d_test
```

```python
d_preds=pd.DataFrame(preds)
d_preds
```

```python
mae_xgb=mean_absolute_error(d_test,d_preds)
mae_xgb
```

```python
r2_score_xgb=r2_score(d_test,d_preds)
r2_score_xgb
```

```python
sns.regplot(d_test,d_preds)
plt.xlabel('Actual_marks')
plt.ylabel('Predicted_marks')
```
